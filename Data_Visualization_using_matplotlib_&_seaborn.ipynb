## Data Cleaning

A vital step of data cleaning is removing missing values displayed as 'na' or 'nan' in dataframes. These missing values can cause errors in matplotlib and should be removed before any data visualization is built. Note below that only one row had to be removed due to missing values.

```python
#inspect original length of dataframe
#displays (number of rows, number of columns)
recent_grads.shape
```

## Visualizing the Data

Different visualizations are used to discover certain relationships.

Line graphs are used to show positive, negative, or no correlation between two numerical variables. Below, there is a negative relationship between Class Rank and Median Salary.

```python
#create figure then add ax1 & ax2 subplots
fig, ax = plt.subplots()

#create line plot
plt.plot(recent_grads['Median'], recent_grads['Rank'])
plt.xlabel("Median Salary")
plt.ylabel("Class Rank")
plt.title("Recent Graduate Salary and Rank Trend")
```

A grouped bar plot can help compare two similar variables.

```python
#create a grouped bar plot
#select only Biology & Life Science Majors
bio = recent_grads[recent_grads["Major_category"] == "Biology & Life Science"]
#focus on the first 8 majors listed in Biology and Life Sciences
bio2 = bio[0:8]
#set index for bar plot 
bio2.set_index("Major", inplace=True)
#compare men and women who graduated
bio2[['Men', 'Women']].plot(kind='bar')
#assign title to bar plot
plt.title('Comparing Men & Women Life Science Graduates')
```

Histograms are a great way to view the distribution of a variable in the data set.

```python
#create histogram for Male and Female Biology Majors

#calculate percentage of recent male biology major grads
recent_grads["ShareMen"] = recent_grads["Men"] / recent_grads["Total"]

#create figure and subplots
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

#generate all histograms
ax1.hist(bio['ShareWomen'])
ax2.hist(bio['ShareMen'])

#set titles for all histograms
ax1.set_title('Female Biology Majors')
ax2.set_title('Male Biology Majors')

#set y-axis range for all histograms
ax1.set_ylim(0, 4)
ax2.set_ylim(0, 4)
```

Compare the correlation between Median Salary and Full-Time Employment and the correlation between Median Salary and Part-Time Employment.

```python
#Create scatter plot in ax2 subplot
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

#generate the scatter plot
x = recent_grads['Full_time_year_round']
y = recent_grads['Median']
ax1.scatter(x, y)
#set x-axis label
ax1.set_xlabel("Full-Time Employment")
#set y-axis label
ax1.set_ylabel("Median Salary")

#generate the scatter plot
x = recent_grads['Part_time']
y = recent_grads['Median']
ax2.scatter(x, y)
#set x-axis label
ax2.set_xlabel("Part-Time Employment")
#set y-axis label
ax2.set_ylabel("Median Salary")
```

Box and whisker plots are a good way to spot outliers!

```python
#box-and-whisker plot
#create list for box plot
num_cols = ["Part_time","Full_time_year_round", "Low_wage_jobs"]

#create plots
fig, ax = plt.subplots()
#create box-and-whisker plot
ax.boxplot(recent_grads[num_cols].values)
#set x-axis tick labels
ax.set_xticklabels(num_cols, rotation=90)
```

## Improving the Aesthetics of Data Visualizations

Be mindful of people who have some amount of color blindness. People who have color blindness have a decreased ability to distinguish between certain kinds of colors.

The Color Blind 10 palette contains ten colors that are colorblind friendly.

The numbers represent the RGB values for each color. The RGB color model describes how the three primary colors (red, green, and blue) can be combined in different proportions to form any secondary color.

```python
#read in csv file
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
#assign colorblind friendly colors to a variable
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
#assign a list of strings to a variable
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
#create a figure
fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    #add multiple subplots to the figure
    ax = fig.add_subplot(1,6,sp+1)
    #plot degree data for women
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    #plot degree data for men
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    #remove spines on all sides of each plot
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    #set x-axis limit to the range 1968 - 2011
    ax.set_xlim(1968, 2011)
    #set y-axis limit to the range 0 - 100
    ax.set_ylim(0,100)
    #assign a string in stem_cats list to its plot
    ax.set_title(stem_cats[sp])
    #hide tick marks on plot
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    #insert text on the leftmost plot, numerically designated a '0'
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    #insert text on the rightmost plot, numerically designated a '5'
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')

#display plot
plt.show()
```

Up to this point, matplotlib as been used to generate the data visualization. The seaborn library is built on top of the matplotlib library. Seaborn has good support for more complex plots, attractive default styles, and integrates well with the pandas library.

Now, let's explore what data visualizations can be made with Seaborn!

```python
#import packages
import seaborn as sns    # seaborn is commonly imported as `sns`
```

```python
#read in csv file
titanic = pd.read_csv('train.csv')
#save list of column names
cols = ["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
#load certian columns from titanic csv file
#remove missing values from csv file
titanic = titanic[cols].dropna()
```

The seaborn.set_style() function changes the default seaborn style sheet. 

Seaborn comes with a few style sheets:
darkgrid: Coordinate grid displayed, dark background color
whitegrid: Coordinate grid displayed, white background color
dark: Coordinate grid hidden, dark background color
white: Coordinate grid hidden, white background color
ticks: Coordinate grid hidden, white background color, ticks visible

```python
#generate a grid of plots for:
#"Survived" column across the columns in the grid
#"Pclass" column across the rows in the grid
#"Sex" column using different hues
#make plot 3 inches high
g = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex", height=3)
#hide coordinate grid and set background white
sns.set_style("white")
#area under the curve shaded
g.map(sns.kdeplot, "Age", shade=True)
#set x-axis label to "Age"
plt.xlabel("Age") 
#add legend
g.add_legend()
#despine plots
sns.despine(left=True, bottom=True)
#display plots
plt.show()
```
